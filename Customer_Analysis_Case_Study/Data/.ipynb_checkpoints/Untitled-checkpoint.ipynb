```python
def remove_outlier(column_name, df):
    percentile25 = df[column_name].quantile(0.25)
    percentile75 = df[column_name].quantile(0.75)
    iqr = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    
    df.replace(to_replace = df[column_name][df[column_name] > upper_limit], value = "new", inplace = True)
    df.replace(to_replace = df[column_name][df[column_name] < lower_limit], value = "new", inplace = True)
    print(f'you dropped {df[column_name][df[column_name] > upper_limit].shape[0] + df[column_name][df[column_name] < lower_limit].shape[0]} values')
    return df
#remove_outlier(‘customer lifetime value’, df)
```

```python
def remove_outlier(column_name, df):
    percentile25 = df[column_name].quantile(0.25)
    percentile75 = df[column_name].quantile(0.75)
    iqr = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    df.loc[(df[column_name] > upper_limit) | (df[column_name] < lower_limit), column_name]='new'
    #df.loc[df[column_name] < lower_limit, column_name]='new'
    print(f'you dropped {len(df[df[column_name]=="new"])} values')
    return df
```
