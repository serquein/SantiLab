Aggregate the data into one Data Frame using Pandas.

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv') 
```

```python
#otra forma:
#file1= file1.rename(columns={file1.columns[1]: 'state'})


new_colums_names = {'ST':'state','GENDER':'gender'}

file1 = file1.rename(columns={'ST':'state','GENDER':'gender'})
file2 = file2.rename(columns={'ST':'state','GENDER':'gender'})
file3 = file3.rename(columns={'State':'state','Gender':'gender'})
```

Pay attention that files may have different names for the same column. therefore, make sure that you unify the columns names before concating them.

```python
file4 = pd.concat([file1, file2, file3]) 
file4 
```

Standardizing header names

```python
def lower_case_column_names(file4):
    file4.columns=[i.lower() for i in file4.columns]
    return file4
file4=lower_case_column_names(file4)

file4
```

```python
file4.drop_duplicates()
```

Deleting and rearranging columns: delete the column customer as it is only a unique identifier for each row of data

```python
file4=file4.drop('customer', axis=1, inplace= True)
```

Removing duplicates

```python
file4.drop_duplicates()
```

Working with data types: Check the data types of all the columns and fix the incorrect ones 

(for ex. customer lifetime value and number of open complaints ). 

Hint: remove the percentage from the customer lifetime value and truncate it to an integer value.

```python
#Check the data types of all the columns
file4.info()
```

```python
#cambiar las int(object) en float de las columnas:
#costumer lifetime value & number of open complaints
file4['customer lifetime value'] = pd.to_numeric(file4['customer lifetime value'], errors='coerce')
file4['number of open complaints '] = pd.to_numeric(file4['number of open complaints'], errors='coerce')
```

```python
#find the NaN values
corrupted_rows_indices=list(file4[file4['customer lifetime value'].isnull()].index)
corrupted_rows_indices
file4.iloc[corrupted_rows_indices,:]
```

```python
#
```

```python
#remove the '%'
#convert to an int
#necesito identificar la cantidad de NaN en la columna para evaluar si los borro o no, si vale la pena hacer el cambio.
file4['customer lifetime value'] = file4['customer lifetime value'].str.replace('%', '').astype(int) 

file4
```
