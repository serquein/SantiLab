```python
#Aggregate the data into one Data Frame using Pandas.
import numpy as np
import pandas as pd

file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv') 

#Aggregate the data into one Data Frame using Pandas.
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv') 

file1 = file1.rename(columns={'ST':'state','GENDER':'gender'})
file2 = file2.rename(columns={'ST':'state','GENDER':'gender'})
file3 = file3.rename(columns={'State':'state','Gender':'gender'})

#Pay attention that files may have different names for the same column. therefore, make sure that you unify the columns names before concating them.
file4 = pd.concat([file1, file2, file3]) 

#Standardizing header names
def lower_case_column_names(file4):
    file4.columns=[i.lower() for i in file4.columns]
    return file4
file4=lower_case_column_names(file4)

#Removing duplicates
file4.drop_duplicates()

#Deleting and rearranging columns: delete the column customer as it is only a unique identifier for each row of data

# otra forma: file4 = file4.drop('customer', axis=1)
file4.drop('customer', axis=1, inplace =True)

file4
```
